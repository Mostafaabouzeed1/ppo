# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
diabetes = pd.read_csv("diabetes.csv")

# 1. Simple Diabetes Count (Bar Chart)
plt.figure(figsize=(8,5))
diabetes['Diabetes_012'].value_counts().plot(kind='bar', color=['green','orange','red'])
plt.title('Number of People with Diabetes Status')
plt.xlabel('Diabetes (0=None, 1=Prediabetes, 2=Diabetes)')
plt.ylabel('Count')
plt.show()

# 2. BMI Comparison (Box Plot)
plt.figure(figsize=(8,5))
sns.boxplot(x='Diabetes_012', y='BMI', data=diabetes)
plt.title('BMI Levels by Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('BMI')
plt.show()

# 3. Diabetes by Age (Line Chart)
plt.figure(figsize=(8,5))
diabetes.groupby('Age')['Diabetes_012'].mean().plot()
plt.title('Average Diabetes Risk by Age')
plt.xlabel('Age Group (1-13)')
plt.ylabel('Diabetes Risk')
plt.show()
```
